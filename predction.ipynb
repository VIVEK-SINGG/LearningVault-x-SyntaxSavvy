```python
# Load libraries
from fbprophet import Prophet
import pandas as pd

# Load data
data = pd.read_csv('pollution_data.csv')

# Prepare data for Prophet
df = pd.DataFrame()
df['ds'] = pd.to_datetime(data['year'], format='%Y')
df['y'] = data['deaths_by_pollution']

# Train-test split
train_size = int(len(df) * 0.8)
train_df = df[:train_size]
test_df = df[train_size:]

# Initialize model
model = Prophet()

# Fit model to training data
model.fit(train_df)

# Make predictions for test data
forecast = model.predict(test_df)

# Evaluate performance
mse = ((forecast['yhat'] - test_df['y']) ** 2).mean()

# Make predictions for next 10 years
future = model.make_future_dataframe(periods=10, freq='Y')
forecast = model.predict(future)
```

```python
# Visualize forecasted values
fig = model.plot(forecast)
plt.title('Deaths by Pollution Forecast for Next 10 Years')
plt.xlabel('Year')
plt.ylabel('Deaths by Pollution')
plt.show()
```
